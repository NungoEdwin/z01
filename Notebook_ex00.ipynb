```python


# Create an array of 300 zeros
arr = np.zeros(300)

# Reshape it to (3, 100)
reshaped_arr = arr.reshape(3, 100)

print(reshaped_arr)
```

```python

# Create a 1D array with integers from 1 to 100
arr = np.arange(1, 101)

print(arr)
```

```python

# Select only odd numbers
odd_arr = arr[arr % 2 != 0]

print(odd_arr)
```

```python


# Select only even numbers
even_arr = arr[arr % 2 == 0][::-1]

print(even_arr)
```

```python
# Set every 3rd element starting from index 1 to 0
arr[1::3] = 0

print(arr)
```
