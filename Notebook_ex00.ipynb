```python


# Create an array of 300 zeros
arr = np.zeros(300)

# Reshape it to (3, 100)
reshaped_arr = arr.reshape(3, 100)

print(reshaped_arr)
```

```python

# Create a 1D array with integers from 1 to 100
arr = np.arange(1, 101)

print(arr)
```

```python

# Select only odd numbers
odd_arr = arr[arr % 2 != 0]

print(odd_arr)
```

```python


# Select only even numbers
even_arr = arr[arr % 2 == 0][::-1]

print(even_arr)
```

```python
# Set every 3rd element starting from index 1 to 0
arr[1::3] = 0

print(arr)
```

```python
# 1. Set the seed to 888 for reproducibility
np.random.seed(888)

# 2. Generate a 1D array of size 100 with a normal distribution
#    (mean = 0, std = 1 by default with np.random.randn)
normal_array = np.random.randn(100)

# 3. Generate a 2D array (8x8) with random integers from 1 to 10 (inclusive)
uniform_2d_array = np.random.randint(1, 11, size=(8, 8))  # upper bound is exclusive

# 4. Generate a 3D array (4x2x5) with random integers from 1 to 17 (inclusive)
uniform_3d_array = np.random.randint(1, 18, size=(4, 2, 5))

# Print outputs
print("1D Normal Distribution Array (100 elements):\n", normal_array)
print("\n2D Uniform Integer Array (8x8):\n", uniform_2d_array)
print("\n3D Uniform Integer Array (4x2x5):\n", uniform_3d_array)
```
