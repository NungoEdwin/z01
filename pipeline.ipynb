```python
#Exercise 1
import numpy as np
from sklearn.impute import SimpleImputer


train_data = [[7, 6, 5],
              [4, np.nan, 5],
              [1, 20, 8]]
test_data = [[np.nan, 1, 2],
             [7, np.nan, 9],
             [np.nan, 2, 4]]            

train_data = [[7, 6, 5],
              [4, np.nan, 5],
              [1, 20, 8]]

imputer = SimpleImputer(strategy='mean')
imputer.fit(train_data)
print(imputer.statistics_)

```

```python
train_data_filled = imputer.transform(train_data)
print(train_data_filled)
```

```python
test_data = [[np.nan, 1, 2],
             [7, np.nan, 9],
             [np.nan, 2, 4]]

test_data_filled = imputer.transform(test_data)
print(test_data_filled)
```

```python
#Exercise 2
from sklearn.preprocessing import StandardScaler

X_train = np.array([[ 1., -1.,  2.],
                     [ 2.,  0.,  0.],
                     [ 0.,  1., -1.]])

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
print('Scaled', X_train_scaled)
print('Scaled mean',scaler.mean_)
print('Scaled std',scaler.scale_)

```

```python
X_test = np.array([[ 2., -1.,  1.],
                   [ 3.,  3., -1.],
                   [ 1.,  1.,  1.]])

X_test_scaled = scaler.transform(X_test)
print("Scaled X_test:\n", X_test_scaled)
```

```python
#Exercise 3
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

X_train = [['Python'], ['Java'], ['Java'], ['C++']]

encoder = OneHotEncoder(handle_unknown='ignore')
X_train_encoded = encoder.fit_transform(X_train).toarray()
df = pd.DataFrame(X_train_encoded, columns=encoder.get_feature_names_out())
print(df)

```

```python
X_test = [['Python'], ['Java'], ['C'], ['C++']]
X_test_encoded = encoder.transform(X_test).toarray()
df_test = pd.DataFrame(X_test_encoded, columns=encoder.get_feature_names_out())
print(df_test)
```

```python
#Exercise 4
from sklearn.preprocessing import OrdinalEncoder

X_train = [['good'], ['bad'], ['neutral']]

encoder = OrdinalEncoder(categories=[['bad', 'neutral', 'good']])
X_train_encoded = encoder.fit_transform(X_train)
print("Encoded X_train:\n", X_train_encoded)
print("Categories:", encoder.categories_)

```

```python
X_test = [['good'], ['good'], ['bad']]
X_test_encoded = encoder.transform(X_test)
print("Encoded X_test:\n", X_test_encoded)
```

```python
#Exercise 5
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer

# Load data (assuming file path)
df = pd.read_csv('breast-cancer.csv', header=None, names=['age', 'menopause', 'tumor-size', 'inv-nodes', 'node-caps', 'deg-malig', 'breast', 'breast-quad', 'irradiat', 'Class'])
print(df.columns)
df = df.drop('Class', axis=1).dropna()
X_train, X_test = train_test_split(df, test_size=0.2, random_state=43)

```

```python
print(X_train.nunique())
```

```python
ohe_cols = ['node-caps', 'breast', 'breast-quad', 'irradiat']
ohe = OneHotEncoder(handle_unknown='ignore')
X_train_ohe = ohe.fit_transform(X_train[ohe_cols])
X_test_ohe = ohe.transform(X_test[ohe_cols]).toarray()
print("OHE features:", ohe.get_feature_names_out(ohe_cols))
print(X_test_ohe[:10])
```

```python
ordinal_cols = ["menopause", "age", "tumor-size", "inv-nodes", "deg-malig"]
categories = [
    ['lt40', 'premeno', 'ge40'],
    ['10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-99'],
    ['0-4', '5-9', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59'],
    ['0-2', '3-5', '6-8', '9-11', '12-14', '15-17', '18-20', '21-23', '24-26', '27-29', '30-32', '33-35', '36-39'],
    [1, 2, 3]
]
oe = OrdinalEncoder(categories=categories)
X_train_oe = oe.fit_transform(X_train[ordinal_cols])
X_test_oe = oe.transform(X_test[ordinal_cols])
print(X_test_oe[:10])
```

```python
preprocessor = make_column_transformer(
    (ohe, ohe_cols),
    (oe, ordinal_cols)
)
X_train_combined = preprocessor.fit_transform(X_train)
X_test_combined = preprocessor.transform(X_test)
print(X_test_combined[:2])
```
