```python
#Exercise 1
import numpy as np
from sklearn.impute import SimpleImputer


train_data = [[7, 6, 5],
              [4, np.nan, 5],
              [1, 20, 8]]
test_data = [[np.nan, 1, 2],
             [7, np.nan, 9],
             [np.nan, 2, 4]]            

train_data = [[7, 6, 5],
              [4, np.nan, 5],
              [1, 20, 8]]

imputer = SimpleImputer(strategy='mean')
imputer.fit(train_data)
print(imputer.statistics_)

```

```python
train_data_filled = imputer.transform(train_data)
print(train_data_filled)
```

```python
test_data = [[np.nan, 1, 2],
             [7, np.nan, 9],
             [np.nan, 2, 4]]

test_data_filled = imputer.transform(test_data)
print(test_data_filled)
```

```python
#Exercise 2
from sklearn.preprocessing import StandardScaler

X_train = np.array([[ 1., -1.,  2.],
                     [ 2.,  0.,  0.],
                     [ 0.,  1., -1.]])

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
print('Scaled', X_train_scaled)
print('Scaled mean',scaler.mean_)
print('Scaled std',scaler.scale_)

```

```python
X_test = np.array([[ 2., -1.,  1.],
                   [ 3.,  3., -1.],
                   [ 1.,  1.,  1.]])

X_test_scaled = scaler.transform(X_test)
print("Scaled X_test:\n", X_test_scaled)
```

```python
#Exercise 3
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

X_train = [['Python'], ['Java'], ['Java'], ['C++']]

encoder = OneHotEncoder(handle_unknown='ignore')
X_train_encoded = encoder.fit_transform(X_train).toarray()
df = pd.DataFrame(X_train_encoded, columns=encoder.get_feature_names_out())
print(df)

```

```python
X_test = [['Python'], ['Java'], ['C'], ['C++']]
X_test_encoded = encoder.transform(X_test).toarray()
df_test = pd.DataFrame(X_test_encoded, columns=encoder.get_feature_names_out())
print(df_test)
```

```python
#Exercise 4
from sklearn.preprocessing import OrdinalEncoder

X_train = [['good'], ['bad'], ['neutral']]

encoder = OrdinalEncoder(categories=[['bad', 'neutral', 'good']])
X_train_encoded = encoder.fit_transform(X_train)
print("Encoded X_train:\n", X_train_encoded)
print("Categories:", encoder.categories_)

```
